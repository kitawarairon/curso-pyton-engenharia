````python
nomes = [
         'Pedro', 'Joao', 'Ronaldo',
         'Fernanda', 'Beatriz', 'Julia',
         'Andre', 'Pollyana', 'Cristina'
        ]

idades = [
    22, 15, 42,
    18, 22, 22,
    60, 45, 40,
         ]

sexos = [
    'M', 'M', 'M',
    'F', 'F', 'F',
    'M', 'F', 'F',
        ]
````

* Deseja-se criar uma lista única contendo a informação dos clientes juntas em uma única lista. 

Por exemplo:
```python
['Pedro', '22', 'M']
```

````python
['asds'] + ['pedro', 55]
````

````python
i = 0
clientes = list()

while i < len(nomes):
    pessoa = []
    pessoa += [nomes[i]] + [idades[i]] + [sexos[i]]
    clientes += [pessoa]
    i += 1
    
clientes
````

* Deseja-se saber pela lista principal, isto é, a lista criada no item anterior a porcentagem de mulheres cadastradas.

````python
total_mulheres = 0
for cliente in clientes:
    if cliente[-1] == 'F':
        total_mulheres += 1

print(f'A porcentagem de mulheres é de {total_mulheres / len(clientes):.2%}')
````
