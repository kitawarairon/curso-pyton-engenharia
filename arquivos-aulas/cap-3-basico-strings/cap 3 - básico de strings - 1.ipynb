## O que é uma string ???

String é um conjunto de caracteres.

```python
'Eu sou uma string' # string de letras
'12456' # string de números
'\M▬!BË4Ê♠5' # não sei o q sou, mas sei que sou string
"também posso ser em aspas duplas" # tenho várias formas
str(4) # meu resultado é uma string
'' # eu sou um string vaziaaaaaaa
```
**Strings são imutáveis!**

Além disso, a gente pode acessar numéricamente um caracter dentro de uma string...

> `Oi`

> 'O' -> posição 0

> 'i' -> posição 1

## Exemplos

* Em uma empresa o código dos funcionários têm dois tipos de entrada inicial, isto é, AB ou AC. Assim, crie um programa que valide a entrada de funcionários apenas na zona de trabalho deles.

Exemplo de codigo: **AB4587, AC879, AB8754, AC87954**

````python
# entrada
# codigo

# saida
# entrada liberada ou entrada n liberada
````

````python
codigo = input('Entre com o seu código')
           # AB4587
if 'AB' in codigo:
    print('Entrada liberada para o portão AB!')
elif 'AC' in codigo:
    print('Entrada liberada para o portão AC!')
else:
    print('Código inválido')
````

* Crie um programa que valide uma senha. A senha precisa ter no mínimo 5 caracteres e no mínimo um caracter especial (%, &, @).

````python
senha = input('Senha: ')

if len(senha) > 5:
    if '%' in senha:
        print('Senha ok!')
    elif '&' in senha:
        print('Senha ok!')
    elif '@' in senha:
        print('Senha ok!')
    else:
        print('A senha está faltando um caracter especial!')
else:
    print('A senha está muito curta!')
````
