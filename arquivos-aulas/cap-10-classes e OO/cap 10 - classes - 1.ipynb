````python
class Dog:
    
    def __init__(self, nome, raca, cor, idade):
        self.nome = nome
        self.raca = raca
        self.cor = cor
        self.idade = idade
    
    
    def sentar(self):
        print(f'{self.nome} está sentando!')
        
    
    def correr(self):
        print(f'{self.nome} está correndo!')
    
    
````

````python
tom = Dog(nome='Tom', raca='Shitzsu', cor='preto-branco', idade=7)
````

````python
tom.sentar()
````

````python
tom.correr()
````

````python
marujo = Dog('Marujo', 'Border Colier', 'preto-branco', 0.4)
````

````python
marujo.correr()
````
