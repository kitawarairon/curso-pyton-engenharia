## Entender o que foi pedido
---
Uma empresa de cálculo estrutural necessitava da criação de uma mecanismo de calculasse a armadura londitudinal de vigas de concreto armado. A equipe de calculo estrutural irá ajudar na parte técnica e auxiliar.

## Dimensionamento de armaduras de vigas de concreto armado
---

* O que uma viga de concreto armado ?



* Como dimensionar qualquer coisa na vida ?

Primeiro, você "chuta" uma dimensão e testa se ela vai atender as verificações. Se atender OK! Se não aumenta a área.

* Passo a passo do dimensionamento:
1. Escolha dos materias (aço e resistência do concreto);
2. Escolhe uma seção transversal;
3. Escolhe uma taxa de agressividade do ambiente;
4. Faz o calculo da linha neutra;
5. Verifica a linha neutra;



6. Faz o cálculo da armadura!
