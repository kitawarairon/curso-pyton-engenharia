```python
import numpy as np
import matplotlib.pyplot as plt
```

## Dados de vendas
---

```python
vendas1 = [14524, 12456, 28456, 20142, 18456]
vendas2 = [12478, 11478, 35421, 22145, 20145]
meses = ['janeiro','fevereiro','março','abril','maio']
```

```python
matrix1 = np.array(vendas1)
meses = np.array(meses)
matrix2 = np.array(vendas2)
```

## Criando primeiro gráfico de linha
---
* Sem subplot

```python
plt.figure()
plt.plot(meses, matrix1)
plt.title('Vendas 1')
plt.show()
```

* Com subplot

```python
fig, ax = plt.subplots()
plt.plot(meses, matrix1)
ax.set_title('Vendas 1')
plt.show()
```

## Criando um gráfico comparativo
---
* **Linha:**

 Criando a função seno e cosseno

```python
# limites 
inf = -6
sup = 6

# dominio
x = np.linspace(inf, sup, 1000)
y_sin = 2 * np.sin(x)
y_cos = 3 *np.cos(x)

# criando nossa figura
plt.figure()
plt.grid()
plt.plot(x, y_sin, label='seno', color='darkseagreen')
plt.plot(x, y_cos, label='cosseno', color='teal')
plt.hlines(0, inf, sup, color='black')
plt.vlines(0, inf, sup, color='black')
plt.title('Seno e Cosseno $x$', fontsize=14, color='green')
plt.legend()
plt.show()
```

* **Barra:**

Criando uma comparação entre as vendas

```python
vendas1 = [14524, 12456, 28456, 19142, 18456]
vendas2 = [10478, 10478, 38421, 22145, 20145]
meses = ['janeiro','fevereiro','março','abril','maio']

x = np.arange(len(meses))
width = 0.35

plt.figure(dpi=250)
fig, ax = plt.subplots()
rect1 = ax.bar(x - width/2, vendas1, width=width, label='Vendas 1', color='gray')
rect2 = ax.bar(x + width/2, vendas2, width=width, label='Vendas 2', color='orange')
plt.grid()

ax.set_xlabel('Meses', color='darkorange', fontsize=14)
ax.set_ylabel('Total de vendas', color='darkorange', fontsize=14)
ax.set_title('Comparação entre as vendas 1 e 2', color='darkorange', fontsize=16)
ax.set_xticklabels(['', 'janeiro','fevereiro','março','abril','maio'])
ax.legend()
plt.show()
```

## Criando subplots
---

* subplot

```python
# dominio
x = np.linspace(inf, sup, 1000)
y_sin = 2 * np.sin(x)
y_cos = 3 *np.cos(x)

# criando uam figura
plt.figure(figsize=(6, 8))

# criando o primeiro plot
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Seno', color='navy')

# criando o segundo plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosseno', color='navy')
```

```python
# dominio
x = np.linspace(inf, sup, 1000)
y_sin = 2 * np.sin(x)
y_cos = 3 *np.cos(x)

# criando uam figura
plt.figure(figsize=(6, 8))

# criando o primeiro plot
plt.subplot(2, 2, 1)
plt.plot(x, y_sin)
plt.title('primeiro plot pos 1', color='navy')

# criando o segundo plot
plt.subplot(2, 2, 2)
plt.plot(x, y_sin)
plt.title('primeiro plot pos 2', color='red')


# criando o terceiro plot
plt.subplot(2, 2, 3)
plt.plot(x, y_cos)
plt.title('primeiro plot pos 3', color='navy')

# criando o quarto plot
plt.subplot(2, 2, 4)
plt.plot(x, y_cos)
plt.title('primeiro plot pos 4', color='red')
```

* subplots

```python
fig, ax = plt.subplots()
ax.plot(x, y_sin)
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, y_sin)
ax[0].set_title('linha 0 coluna 0')
ax[1].plot(x, y_cos)
ax[1].set_title('linha 1 coluna 0')
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, y_sin)
ax[0].set_title('linha 0 coluna 0')
ax[1].plot(x, y_cos)
ax[1].set_title('linha 1 coluna 0')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y_sin)
ax[0, 0].set_title('linha 0 coluna 0')

ax[0, 1].plot(x, y_sin)
ax[0, 1].set_title('linha 0 coluna 1')


ax[1, 0].plot(x, y_cos)
ax[1, 0].set_title('linha 1 coluna 0')

ax[1, 1].plot(x, y_cos)
ax[1, 1].set_title('linha 1 coluna 1')




fig.tight_layout()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)


fig, axs = plt.subplots(3, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'o')
axs[2].plot(x, y, '+')
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)


fig, axs = plt.subplots(3, sharex=False, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'o')
axs[2].plot(x, y, '+')
```

```python
fig = plt.figure()
gs = fig.add_gridspec(3, hspace=0)
axs = gs.subplots(sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'o')
axs[2].plot(x, y, '+')

# Hide x labels and tick labels for all but bottom plot.
for ax in axs:
    ax.label_outer()
```

## Cores
---

https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, title, sort_colors=True, emptycols=0):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12
    topmargin = 40

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in colors.items())
        names = [name for hsv, name in by_hsv]
    else:
        names = list(colors)

    n = len(names)
    ncols = 4 - emptycols
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + margin + topmargin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-topmargin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()
    ax.set_title(title, fontsize=24, loc="left", pad=10)

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig

plot_colortable(mcolors.BASE_COLORS, "Base Colors",
                sort_colors=False, emptycols=1)
plot_colortable(mcolors.TABLEAU_COLORS, "Tableau Palette",
                sort_colors=False, emptycols=2)

plot_colortable(mcolors.CSS4_COLORS, "CSS Colors")

# Optionally plot the XKCD colors (Caution: will produce large figure)
# xkcd_fig = plot_colortable(mcolors.XKCD_COLORS, "XKCD Colors")
# xkcd_fig.savefig("XKCD_Colors.png")

plt.show()
```
