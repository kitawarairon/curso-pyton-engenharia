* Vamos criar uma função para calcular a raiz do df

 nós podemos definir nossa própria função; <br>

 .apply() -> recebe o todo dataframe, caso a gente não específique, ela irá aplicar em todos os elementos ou dará algum erro;

```python
novo_df.apply(np.sqrt).head(3)                         # utilizei a função do NumPY
                                                       # a função está aplicando sob todos os elementos do df
```

```python
def raiz(a): return (a)**(0.5)                         # criando minha function
```

```python
novo_df.apply(raiz).head(3)                            # utilizei com a minha function
```

```python
def raiz3_n1(c):
    if c['n1']:
        return c['n1']**(1/3)
    
```

```python
novo_df.apply(raiz3_n1, axis=1).head(3)                # apliquei sob uma colunas, olhe a lógica de formação da função
```

> Exemplo desafio: Crie uma function para aplicar no df onde o returno será o maior valor entre as colunas n1 e n2

```python
def my_func(a, n1='n1',n2='n2'):
    if a[n1] >= a[n2]:
        return a[n1]
    else:
        return a[n2]
    
    
```

```python
novo_df.head(3)
```

```python
novo_df['coluna my func'] = novo_df.apply(my_func, axis=1).head(3)   
```

```python
novo_df.head(3)
```
