```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../dados/countries.csv')
df.head()
```

### Uma análise entre Brasil x Argentina;

```python
brasil = df[df.country == 'Brazil']
brasil
```

```python
argentina = df[df.country == 'Argentina']
argentina
```

```python
x_br = brasil.year
y_br = brasil.population

x_arg = argentina.year
y_arg = argentina.population



# vamos criar nossa figura
plt.figure()
# plot-brasil
plt.plot(x_br, y_br, ls=':', color='navy', label='População BR')
# plot-argentina
plt.plot(x_arg, y_arg, ls='--', color='red', label='População ARG')
```

```python
# criando os eixos
x_br = brasil.year
y_br = brasil.population
x_arg = argentina.year
y_arg = argentina.population

# vamos criar nossa figura
plt.figure(figsize=(8, 5), dpi=150)
# plot-brasil
plt.plot(x_br, y_br, ls=':', color='navy', label='População BR')
# plot-argentina
plt.plot(x_arg, y_arg, ls='--', color='red', label='População ARG')
# melhorando o nosso gráfico
plt.title('Comparação entre o crescimento da população Brasil x Argentina', fontsize=16, color='navy')
plt.xlabel('Anos', fontsize=14, color='navy')
plt.ylabel('População (em milhões)', color='navy')
plt.grid(True)
plt.legend()
```

### Encontrar os 5 maiores populações;

```python
pop_2007 = df[df.year == 2007]
pop_2007
```

```python
pop_5_maiores_2007 = pop_2007.sort_values(by=['population'], ascending=False).head()
pop_5_maiores_2007
```

```python
# vamos criar nossa figura
plt.figure(figsize=(8, 5), dpi=150)
# plot-brasil
plt.bar(x=pop_5_maiores_2007.country, height=pop_5_maiores_2007.population)

# melhorando o nosso gráfico
plt.title('As cinco maiores populações em 2007', fontsize=14, color='navy')
plt.xlabel('Anos', fontsize=14, color='navy')
plt.ylabel('População (em milhões)', color='navy')
plt.grid(True)
```

### Análise do maior país em população;

```python
china = df[df.country == 'China']
china
```

```python
# criando os eixos
x_chi = china.year
y_chi = china.population

# vamos criar nossa figura
plt.figure(figsize=(8, 5), dpi=150)
# plot-brasil
plt.plot(x_chi, y_chi, ls=':', color='navy', label='População CHI')

# melhorando o nosso gráfico
plt.title('Crescimento da população da China', fontsize=16, color='navy')
plt.xlabel('Anos', fontsize=14, color='navy')
plt.ylabel('População (em milhões)', color='navy')
plt.grid(True)
plt.legend()
```
